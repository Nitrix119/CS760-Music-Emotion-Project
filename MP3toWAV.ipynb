```python
from pydub import AudioSegment
import os
import pandas as pd

# Note: This requires FFmpeg to work due to pydub running on it. 
# Download it and add ffmpeg/bin to PATH and reload the IDE if you are getting errors about ffprobe being missing.
```

```python
#Load Data
df = pd.read_csv('multimodal_dataset_normalized.csv')
display(df.head())
```

```python
#Emotion Class
unique_items = df["Quadrant"].unique()
print(unique_items)
```

```python
#Extract path and emotion
data_dir = "MERGE_Bimodal_Complete/audio"
labels_dict = {f"{row['Quadrant']}/{row['Audio_Song']}.mp3": row['Quadrant']
               for _, row in df.iterrows()}
print(labels_dict)
```

```python
def convert_mp3_to_wav(mp3_path, wav_path, target_sr=16000):
    audio = AudioSegment.from_mp3(mp3_path)
    audio = audio.set_frame_rate(target_sr).set_channels(1)
    audio.export(wav_path, format="wav")

# Example usage
wav_dir = "MERGE_Bimodal_Complete/audio_wav"
to_do = len(labels_dict)
done = 0
for mp3, label in labels_dict.items():
    mp3_path = os.path.join(data_dir, mp3)
    wav_path = mp3_path.replace(data_dir, wav_dir).replace(".mp3", ".wav")
    convert_mp3_to_wav(mp3_path, wav_path)
    
    done += 1
    print(f"Completed {done}/{to_do}", end="\r")
```
